```python
from sklearn.externals import joblib
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from scipy.interpolate import interp1d
import os
from os import walk
import time
```

```python
prev_count = 0
comp_count = 0
results = []
```

```python
mypath = os.getcwd() + "\\predict_test\\"
while(True):
    count = 0
    with open(mypath + "capture_test.txt") as f:
        for line in f:
            count += 1
    if(count > comp_count):
        time.sleep(3)
        with open(mypath + 'out.txt') as f:
            for line in f :
                prev_count += 1

        file = open(mypath + "out.txt","r+")        
        file.truncate(0)

        curr_count = 0
        with open(mypath + "capture_test.txt") as f:
            with open(mypath + "out.txt", "w") as f1:
                for line in f:
                    curr_count += 1
                    if curr_count > prev_count:    
                        f1.write(line)
        comp_count = curr_count
        
        data = pd.DataFrame()
        #for file in f:
        a = pd.read_csv(mypath + "out.txt", sep = ' ', header = None)
        del a[0]
        del a[7]
        a = np.array(a).astype(float)
        data_norm = scale(a)

        acx = data_norm[:,0]
        acy = data_norm[:,1]
        acz = data_norm[:,2]
        gx = data_norm[:,3]
        gy = data_norm[:,4]
        gz = data_norm[:,5]

        x = np.linspace(0, a.shape[0], a.shape[0])
        f_acx = interp1d(x, acx)
        f_acy = interp1d(x, acy)
        f_acz = interp1d(x, acz)
        f_gx = interp1d(x, gx)
        f_gy = interp1d(x, gy)
        f_gz = interp1d(x, gz)

        xnew = np.linspace(0, a.shape[0], 50)
        acx_stretch = f_acx(xnew)
        acy_stretch = f_acy(xnew)
        acz_stretch = f_acz(xnew)
        gx_stretch = f_gx(xnew)
        gy_stretch = f_gy(xnew)
        gz_stretch = f_gz(xnew)

        xxx = np.concatenate((acx_stretch, acy_stretch, acz_stretch, gx_stretch, gy_stretch, gz_stretch))
        xxx = pd.DataFrame(xxx.reshape(-1, len(xxx)))
        data = data.append(xxx)

        data.reset_index(drop = True, inplace = True)
        
        clf_best = joblib.load('best_model.pkl')
        clf1 = joblib.load('lr_model.pkl')
        clf2 = joblib.load('lsvm_model.pkl')
        clf3 = joblib.load('ksvm_model.pkl')
        clf4 = joblib.load('knn_model.pkl')
        clf5 = joblib.load('nb_model.pkl')
        clf6 = joblib.load('dt_model.pkl')
        clf7 = joblib.load('rf_model.pkl')
        
        acc_list = []
        pred_clf_best = clf_best.predict(data)
        acc_list.append(pred_clf_best)
        pred_clf1 = clf1.predict(data)
        acc_list.append(pred_clf1)
        pred_clf2 = clf2.predict(data)
        acc_list.append(pred_clf2)
        pred_clf3 = clf3.predict(data)
        acc_list.append(pred_clf3)
        pred_clf4 = clf4.predict(data)
        acc_list.append(pred_clf4)
        pred_clf5 = clf5.predict(data)
        acc_list.append(pred_clf5)
        pred_clf6 = clf6.predict(data)
        acc_list.append(pred_clf6)
        pred_clf7 = clf7.predict(data)
        acc_list.append(pred_clf7)
        
        print('The predicted letter from the models are :- ')
        print("Logistic Regression - ",acc_list[0][0])
        print("Linear SVM - ",acc_list[1][0])
        print("Kernel SVM - ",acc_list[2][0])
        print("K-Nearest Neighbors - ",acc_list[3][0])
        print("Naive Bayes - ",acc_list[4][0])
        print("Decision Tree - ",acc_list[5][0])
        print("Random Forest - ",acc_list[6][0])

        
        results.append(pred_clf_best[0])
        with open(mypath + "onscreen.txt","a") as f:
            if pred_clf_best[0]=="sp":
                f.write(" ")
            elif pred_clf_best[0]=="en":
                f.write('\n')
            else:
                f.write(pred_clf_best[0])
        print("")
        print("-----------------------------------------------------")
        print("")
```
